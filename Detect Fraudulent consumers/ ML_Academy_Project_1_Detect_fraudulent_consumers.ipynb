# **Importing Data**

```python
import pandas as pd

path_train = '/content/drive/MyDrive/ML Academy - Project 1/STEG Dataset/Train'
path_test = '/content/drive/MyDrive/ML Academy - Project 1/STEG Dataset/Test'
# Read the files into two dataframes.
df1 = pd.read_csv(path_train + '/client_train.csv')
df2 = pd.read_csv(path_train + '/invoice_train.csv')
```

```python
# client_train
df1
```

```python
# invoice_train
df2
```

```python
# Merge the two dataframes, using _ID column as key
df3 = None
df3 = pd.merge(df2,df1, on = 'client_id')
df3.set_index('client_id', inplace = True)
```

```python
df3
```

```python
merged_dataset = df3.copy()
merged_dataset
```

# **Missing Values**

```python
# Finding Missing Values :
merged_dataset.isna().sum() 
```

# **Duplicate Rows**

```python
# Total number of rows and columns
print(merged_dataset.shape)

# Rows containing duplicate data
duplicate_rows_df = merged_dataset[merged_dataset.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
# keep first duplicate row
Duplicated_dataset = merged_dataset.drop_duplicates() 
print('Result DataFrame:\n', Duplicated_dataset)
Duplicated_dataset
```

```python
Duplicated_dataset.info()
```

```python
Duplicated_dataset.describe().transpose()
```
